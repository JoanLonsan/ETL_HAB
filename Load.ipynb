## Subir DF a AirTable
### at_upload_df(df, base_id, table_id, token)

```python
def at_upload_df(df, base_id, table_id, token):
    """
    Sube un DataFrame a una tabla de Airtable.
    Esta función toma un DataFrame y sube sus registros como entradas a una tabla en Airtable.
    Args:
        df (pandas.DataFrame): El DataFrame a subir.
        base_id (str): ID de la base de datos en Airtable.
        table_id (str): ID de la tabla en Airtable.
        token (str): Token de autorización de Airtable.
    Returns:
        None
    """
    registros = []
    for _, row in df.iterrows():
        registro = {}
        registro["fields"] = {}
        for col, value in row.items():
            registro["fields"][col] = value
        registros.append(registro)
    
    table_data = {"records": registros, "typecast": True}
    
    airtable_base_url = "https://api.airtable.com/v0"
    endpoint = f"{airtable_base_url}/{base_id}/{table_id}"
    
    headers = {"Authorization" : f"Bearer {token}",
               "Content-Type"  : "application/json"}
    
    for i in range(0, len(table_data["records"]), 10):
        data = {"records": table_data["records"][i : i + 10]}
        response_post = post(url=endpoint, json=data, headers=headers)
        sleep(0.5)
```

## Descargar DF de AirTable

```python
def df_order(df):
    """
    Reordena las columnas del DataFrame especificado.
    Esta función reorganiza las columnas del DataFrame en un orden específico para
    mejorar la legibilidad o facilitar ciertas operaciones.
    Args:
        df (pandas.DataFrame): El DataFrame a reorganizar.
    Returns:
        pandas.DataFrame: El DataFrame con las columnas reordenadas.
    """
    df = df[['Pais', 'Orden en el Top 50', 'Nombre de la Cancion', 'id', 'Popularidad', 'Artista',
         'danceability', 'energy', 'loudness', 'mode', 'speechiness',
         'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo',
         'duration_ms', 'duration_mins', 'time_signature','key', 'key_musical']]
    return df
```

```python
def df_order_2(df):
    """
    Reordena las columnas del DataFrame especificado.
    Esta función reorganiza las columnas del DataFrame en un orden específico para
    mejorar la legibilidad o facilitar ciertas operaciones.
    Args:
        df (pandas.DataFrame): El DataFrame a reorganizar.
    Returns:
        pandas.DataFrame: El DataFrame con las columnas reordenadas.
    """
    df = df[['artists_top10', 'position_top10', 'genre', 'followers_top10', 'track_name', 'track_position', 'album', 'release_date', 'id',
            'danceability', 'energy', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness',
            'liveness', 'valence', 'tempo', 'duration_ms', 'duration_mins', 'time_signature','key', 'key_musical']]
    return df
```

```python
def at_download_df(base_id, table_id, token):
    """
    Descarga una tabla de Airtable como un DataFrame de Pandas.
    Esta función realiza una solicitud a la API de Airtable para descargar los registros 
    de una tabla específica y luego los convierte en un DataFrame de Pandas.
    Args:
        base_id (str): ID de la base de datos en Airtable.
        table_id (str): ID de la tabla en Airtable.
        token (str): Token de autorización de Airtable.
    Returns:
        pandas.DataFrame: El DataFrame generado a partir de los registros descargados.
    """
    params = {}
    airtable_base_url = "https://api.airtable.com/v0"
    endpoint = f"{airtable_base_url}/{base_id}/{table_id}"
    headers = {"Authorization" : f"Bearer {TOKEN}",
           "Content-Type"  : "application/json"}
    datos = list()
    while params.get("offset") != None or len(datos) == 0:
        response = get(url = endpoint, headers = headers, params = params)
        print(f"response: {response.status_code}")
        print(f"endpoint: {response.url}")
        data = response.json()
        offset = data.get("offset")
        params["offset"] = offset
        datos.extend(data["records"])
    df = pd.DataFrame(datos)
    df_normalized = pd.json_normalize(df['fields'])
    if table_id == TABLE_TOP50_ID:
        df_normalized = df_order(df_normalized)
    else:
        df_normalized = df_order_2(df_normalized)
    return df_normalized
```
